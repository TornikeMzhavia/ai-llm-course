```python
import ollama

# List available models
response = ollama.list()
print(response)

[print(model["model"]) for model in response["models"]]
```

## Chat example


```python
res = ollama.chat(
    model="llama3.2",
    messages=[
        {"role": "user", "content": "why is the sky blue?"},
    ],
)
print(res["message"]["content"])
```

## Chat example (streaming)


```python
res = ollama.chat(
    model="llama3.2",
    messages=[
        {
            "role": "user",
            "content": "why is the ocean so salty?",
        },
    ],
    stream=True,
)
for chunk in res:
    print(chunk["message"]["content"], end="", flush=True)
```

## Generate example


```python
res = ollama.generate(
    model="llama3.2",
    prompt="why is the sky blue?",
)
print(res["response"])
```

## Show model information


```python
print(ollama.show("llama3.2"))
```

## Create a new model with modelfile


```python
system_prompt = """
You are very smart assistant who knows everything about oceans. You are very succinct and informative.
"""

ollama.create(model="knowitall", from_= "llama3.2", system=system_prompt,parameters={"temperature": 0.1})
```

```python
res = ollama.generate(model="knowitall", prompt="why is the ocean so salty?")
print(res["response"])
```
